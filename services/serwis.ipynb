# Mikroserwis


Mikroserwis należy uruchomić z poziomu folderu services za pomoca komendy: uvicorn main:app --reload;
Przed odpaleniem serwisu należy zainstalować pakiety zawarte w requirements.txt


Z funkcjonalności oferowanych przez serwis można skorzystać za pomocą Swaggera (przechodząc na adres /docs), bądź wykonując requesty z poziomu notatnika. Poniżej prezentujemy wyniki zapytań z poziomy swaggera


Dokonanie predykcji za pomocą modelu bazowego


_moedel bazowy zwraca jeden gatunek muzyczny, na podstawie wprowadzonych parametrów. Parametry w pierwszym kroku zostają normalizowane, a zwraca wartość dekodowana, do czego wykorzystujemy kolejno zserializowany scaler oraz encoder_

Dokonanie predykcji za pomocą modelu docelowego


_Model docedowy jest implementacją sieci neuronowej, wobec czego wynikiem jego predykcji jest wektor prawdopodobieństw, z którego można odczytać kilka najbardziej prawdopodobnych gatunków utworu_
