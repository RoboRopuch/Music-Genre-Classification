# Przygotowanie danych do modelowania

Wczytanie wszystkich danych jako obiekt DataFrame:

```python
import pandas as pd

PATH_ARTISTS = "data/v2/artists.jsonl"
artists_df = pd.read_json(PATH_ARTISTS, lines=True)
artists_df
```

```python
PATH_TRACKS = "data/v2/tracks.jsonl"
tracks_df = pd.read_json(PATH_TRACKS, lines=True)
tracks_df
```

### Grupowanie gatunków na podstawie permutacji
Aby zmniejszyć ilość gatunków, będziemy agregować podobne gatunki do jednej kategorii (np. *greek drill* i *turkish drill* do *drill*).

Najpierw, dane dotyczące gatunków zostaną wczytane jako *MultiLabelBinarizer*

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
genres_df = pd.DataFrame(mlb.fit_transform(artists_df['genres']),
                         columns=mlb.classes_,
                         index=artists_df["id"])
genres_df
```

Znalezione zostaną wszystkie możliwe permutacje (o zmiennej długości) nazw gatunków muzycznych. Jeśli jedna z tych permutacji jest nazwą innego gatunku, gatunki te zostaną połączone.

```python
from itertools import permutations

all_permutations = {}
for genre in genres_df.keys():
    all_permutations.setdefault(genre, [])
    split_genre = genre.split()
    for i in range(1, len(split_genre) + 1):
        length_permutations = permutations(split_genre, i)
        all_permutations[genre].extend([" ".join(genre) for genre in length_permutations])
```

```python
genre_mappings = {}
for original_genre, genre_permutations in all_permutations.items():
    genre_mappings.setdefault(original_genre, [])
    for genre in genres_df.keys():
        if any(permutation == genre for permutation in genre_permutations) and len(genre) < len(original_genre):
            genre_mappings[original_genre].append(genre)

list(genre_mappings.items())[:15]
```

Wszyscy artyści ze zbioru danych *genres_df* mają dodane nowe gatunki, zgodnie z powyższym mapowaniem:

```python
for old_genre, new_genres in genre_mappings.items():
    genres_df.loc[genres_df[old_genre] == 1, new_genres] = 1
```

Następnie te gatunki, które mogły zostać zastąpione innym, bardziej ogólnym, zostaną usunięte z *genres_df*:

```python
replaced_genres = [genre for genre in genre_mappings if genre_mappings[genre]]
genres_df.drop(replaced_genres, axis=1, inplace=True)
genres_df
```

Ilość gatunków została zmniejszona z 3953 do 1586.

Są wciąż pewne często powtarzające się elementy w nazwach gatunków. Są to m.in: indie, dance, traditional, instrumental, electronic, children, folklore, pop. Zostaną one zamienione na bardziej ogólny opis.

```python
repeating_genres = ["indie", "dance", "traditional", "instrumental", "electronic", "children", "folklore", "pop"]
for genre in repeating_genres:
    keys = [k for k in genres_df.keys() if (genre in k and genre != k)]
    if genre not in genres_df.keys():
        genres_df[genre] = 0
    genres_df.loc[genres_df[keys].eq(1).any(axis=1), genre] = 1
    genres_df.drop(keys, axis=1, inplace=True)
genres_df
```

Kod sprawdza, czy żaden artysta nie został bez gatunku:

```python
genres_df[genres_df.sum(axis=1, numeric_only=True) == 0]
```

Następnie należy zmienić dane w tabeli *artists_df*, tak aby odwzorowywał zmienione gatunki muzyczne

```python
genres_dict = genres_df.apply(lambda row: row[row == 1].index.tolist(), axis=1).to_dict()
genres_list_df = pd.DataFrame(list(genres_dict.items()), columns=["id_artist", "genres"])
artists_df = pd.merge(artists_df, genres_list_df, left_on="id", right_on="id_artist", suffixes=("_old", "_new")).drop("id_artist", axis=1)
artists_df
```

W tabeli *tracks_df* należy zamienić kolumnę release_date na release_year, aby wszystkie dane były spójne:

```python
tracks_df["release_year"] = tracks_df["release_date"].str.extract(r'(\d{4})')
tracks_df['release_year'] = pd.to_numeric(tracks_df['release_year'], errors='coerce')
tracks_df
```

Gatunki zostaną zmapowane na piosenki, zgodnie z przypisanym artystą.

```python
genres_track_df = pd.merge(genres_df, tracks_df[["id_artist", "id"]], left_on="id", right_on="id_artist").drop(["id_artist"], axis=1)
genres_track_df = genres_track_df.set_index("id")
genres_track_df
```

### Grupowanie gatunków mało popularnych za pomocą K-means

Jeśli weźmiemy pod uwagę tylko pierwsze 20 najpopularniejszych gatunków, w naszym zbiorze danych pozostanie 23 052 piosenek bez przypisanego gatunku:

```python
popular_genres_list = genres_track_df.sum(numeric_only=True).sort_values(ascending=False)[:20].keys()
new_genres_df = genres_track_df[popular_genres_list]
new_genres_df[new_genres_df.sum(axis=1, numeric_only=True) == 0]
```

Jest to znacząca ilość dla naszego zbioru danych (18%). Aby nie pozbywać się tych danych, ani nie klasyfikować wszystkich pozostałych gatunków jako "inne", wykorzystamy algorytm K-means. Podzieli on pozostałe piosenki na klastry w zależności od wartości atrybutów piosenek. Na podstawie tych klastrów wyłonione zostaną odpowiednie gatunki.

Najpierw dane dotyczące piosenek zostaną znormalizowane:

```python
ids = new_genres_df[new_genres_df.sum(axis=1, numeric_only=True) == 0].index
X = tracks_df[tracks_df['id'].isin(ids)].drop(columns=["mode", "name", "release_date", "energy"])
X = X.set_index("id")
X
```

```python
from sklearn.preprocessing import StandardScaler

features_to_normalize = ['popularity', 'duration_ms', 'danceability', 'key', 'loudness', 'speechiness', 
                         'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                         'release_year', 'time_signature']

scaler = StandardScaler()

X[features_to_normalize] = scaler.fit_transform(X[features_to_normalize])
X
```

*scaler* zostanie zapisany do pliku:

```python
import pickle

with open("./models/artifacts/scalers/scaler.pkl", 'wb') as file:
    pickle.dump(scaler, file)
```

Piosenki będziemy dzielić na 50 klastrów na podstawie znormalizowanych danych piosenek.

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=50, n_init=10, random_state=42)
k_means.fit(X.drop(['id_artist'], axis=1))
X['cluster'] = k_means.predict(X.drop(['id_artist'], axis=1))
X.to_csv('./data/tracks_normalized_params.csv', index=False)

X
```

Do danych piosenek dołączone zostają listy gatunków, które reprezentują.

```python
merged_df = pd.merge(X, genres_list_df, on='id_artist', how='left').set_index(X.index)
merged_df
```

Możemy zobaczyć, jakie gatunki są w każdym klastrze.

```python
genre_counts = {}
for cluster, genres in merged_df.groupby('cluster')['genres']:
    all_genres = [genre for sublist in genres for genre in sublist]
    genre_counts[cluster] = all_genres

genre_counts
```

Każdy klaster będzie miał przypisany jeden gatunek na podstawie częstości występowania - wybrany zostanie najpopularniejszy gatunek w każdym klastrze.

```python
from collections import Counter
genre_counts = {}
for cluster, genres in merged_df.groupby('cluster')['genres']:
    all_genres = [genre for sublist in genres for genre in sublist]
    genre_counts[cluster] = Counter(all_genres)

most_popular_genres = {}
for cluster, counts in genre_counts.items():
    most_popular_genres[cluster] = counts.most_common(1)[0][0]

print(most_popular_genres)
# Extract unique most popular genres
unique_popular_genres = set(most_popular_genres.values())
print("\nUnique Most Popular Genres:")
print(unique_popular_genres)
```

Zamienimy kategorie wg. klastrów - w zależności od najpopularniejszego występującego w nim gatunku.

```python
for cluster, genre in most_popular_genres.items():
    merged_df.loc[merged_df['cluster'] == cluster, "genres"] = genre

merged_df
```

Do zbioru 20 najpopularniejszych gatunków dodajemy gatunki wyłonione w procesie klasteryzacji.

```python
for genre in unique_popular_genres:
    new_genres_df[genre] = 0
    indices = merged_df[merged_df["genres"] == genre].index
    new_genres_df.loc[indices, genre] = 1
new_genres_df
```

Możemy zbadać popularność wszystkich gatunków.

```python
new_genres_df.sum().sort_values(ascending=True)
```

Pięć najmniej popularnych gatunków ma na tyle mało reprezentantów, że trudno byłoby modelowi nauczyć się je rozpoznawać. Wobec tego zostaną zamienione na kategorię "other":

```python
unpopular_genres = new_genres_df.sum().sort_values(ascending=True).keys()[:5]
new_genres_df["other"] = 0
for genre in unpopular_genres:
    new_genres_df.loc[new_genres_df[genre] == 1, "other"] = 1
    new_genres_df.drop(genre, axis=1, inplace=True)

new_genres_df
```
